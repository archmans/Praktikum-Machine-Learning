## Dataset
`diabetes.csv` merupakan dataset yang telah dimodifikasi dari [Diabetes Health Indicators Dataset](https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset/) sebagai kumpulan indikator individu yang diperoleh dari survei untuk kasus diabetes. Dataset ini berguna untuk melakukan prediksi diabetes sehingga suatu individu dapat diketahui memiliki risiko tinggi diabetes atau tidak. Hal ini diperoleh dari fitur-fitur yang dapat dianalisis lebih lanjut sebelum mencapai kesimpulan.

Berikut adalah deskripsi singkat setiap kolom:

1. **HighBP**: Memiliki tekanan darah tinggi (BP: Blood Pressure) atau tidak
2. **HighChol**: Kolesterol tinggi atau tidak
3. **BMI**: Besaran Body Mass Index
4. **Smoker**: Perokok atau bukan perokok
5. **Stroke**: Pernah mengalami struk atau tidak
6. **HeartDiseaseorAttack**: Memiliki riwayat penyakit antara jantung koroner dan serangan jantung atau tidak sama sekali
7. **PhysActivity**: Aktif secara fisik dalam 30 hari terakhir atau tidak
8. **Fruits**: Mengonsumsi buah setiap hari atau tidak 
9. **Veggies**: Mengonsumsi sayur setiap hari atau tidak
10. **HvyAlcoholConsump**: Peminum berat alkohol atau bukan 
11. **AnyHealthcare**: Memiliki perlindungan kesehatan atau tidak, contohnya memiliki asuransi kesehatan
12. **GenHtlth**: Evaluasi mandiri terhadap kesehatan, skala 1-5 (1: Sangat baik, 2: Cukup Baik, 3: Baik, 4: Biasa saja, 5: Buruk)
13. **MentHlth**: Jumlah hari keadaan mental buruk dalam 30 hari terakhir (skala 0-30 hari)  
14. **PhysHlth**: Jumlah hari keadaan fisik buruk dalam 30 hari terakhir (skala 0-30 hari)
15. **DiffWalk**: Memiliki kesulitan berjalan atau menaiki tangga
16. **Sex**: (M) Male atau (F) Female
17. **Age**: 13 kategori umur (1: 18-24 tahun, 9: 60-64 tahun, 13: 80 tahun ke atas)
18. **Education**: Level edukasi skala 1-6 (1: Tidak pernah sekolah atau hanya TK, 2: SD, dst)
19. **Income**: Skala pendapatan 1-8
20. **Diabetes**: Apakah mengalami diabetes atau tidak (Kolom target)

## 1. Baseline dengan menggunakan model logistic regression.

```python
X = data.drop(columns="Diabetes")
y = data["Diabetes"].copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=123)

# Gunakan data validasi untuk bereksperimen dengan model
# Gunakan data test untuk mengevaluasi model hanya di akhir eksperimen
df_train = pd.concat([X_train, y_train], axis=1)
df_val = pd.concat([X_val, y_val], axis=1)
df_test = pd.concat([X_test, y_test], axis=1)
```

```python
logistic_regression = LogisticRegression(max_iter=10000)
logistic_regression.fit(X_train, y_train)
y_pred = logistic_regression.predict(X_val)

accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

## 6. Implementasi strategi eksperimen dan skema validasi yang telah ditentukan pada poin 5.

```python
# Drop duplicates data
train_duplicates_clean = df_train.drop_duplicates()
```

```python
# Replace outlier with median
# Data Pencilan
median_BMI = train_duplicates_clean['BMI'].median()
median_MentHlth = train_duplicates_clean['MentHlth'].median()
median_PhysHlth = train_duplicates_clean['PhysHlth'].median()
data_outliner_cleaned = train_duplicates_clean
print("Missing values:")
print(data_outliner_cleaned.isnull().sum())

Q1_BMI = data_outliner_cleaned['BMI'].quantile(0.25)
Q3_BMI = data_outliner_cleaned['BMI'].quantile(0.75)
IQR_BMI = Q3_BMI - Q1_BMI
lower_bound_BMI = Q1_BMI - 1.5 * IQR_BMI
upper_bound_BMI = Q3_BMI + 1.5 * IQR_BMI

Q1_MentHlth = data_outliner_cleaned['MentHlth'].quantile(0.25)
Q3_MentHlth = data_outliner_cleaned['MentHlth'].quantile(0.75)
IQR_MentHlth = Q3_MentHlth - Q1_MentHlth
lower_bound_MentHlth = Q1_MentHlth - 1.5 * IQR_MentHlth
upper_bound_MentHlth = Q3_MentHlth + 1.5 * IQR_MentHlth

Q1_PhysHlth = data_outliner_cleaned['PhysHlth'].quantile(0.25)
Q3_PhysHlth = data_outliner_cleaned['PhysHlth'].quantile(0.75)
IQR_PhysHlth = Q3_PhysHlth - Q1_PhysHlth
lower_bound_PhysHlth = Q1_PhysHlth - 1.5 * IQR_PhysHlth
upper_bound_PhysHlth = Q3_PhysHlth + 1.5 * IQR_PhysHlth

# Replace outlier with median
data_outliner_cleaned.loc[data_outliner_cleaned['BMI'] < lower_bound_BMI, 'BMI'] = median_BMI
data_outliner_cleaned.loc[data_outliner_cleaned['BMI'] > upper_bound_BMI, 'BMI'] = median_BMI
data_outliner_cleaned.loc[data_outliner_cleaned['MentHlth'] < lower_bound_MentHlth, 'MentHlth'] = median_MentHlth
data_outliner_cleaned.loc[data_outliner_cleaned['MentHlth'] > upper_bound_MentHlth, 'MentHlth'] = median_MentHlth
data_outliner_cleaned.loc[data_outliner_cleaned['PhysHlth'] < lower_bound_PhysHlth, 'PhysHlth'] = median_PhysHlth
data_outliner_cleaned.loc[data_outliner_cleaned['PhysHlth'] > upper_bound_PhysHlth, 'PhysHlth'] = median_PhysHlth

print("data_outliner_cleaned setelah outlier direplace:")
print(data_outliner_cleaned.shape)
```

```python
# Undersampling false data in data_outliner_cleaned
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=123)
X_resampled_under, y_resampled_under = rus.fit_resample(data_outliner_cleaned.drop(columns="Diabetes"), data_outliner_cleaned["Diabetes"])

print("Setelah undersampling:")

# Train model with undersampled data
logistic_regression = LogisticRegression(max_iter=10000)
logistic_regression.fit(X_resampled_under, y_resampled_under)
y_pred = logistic_regression.predict(X_val)

accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
# Oversampling true data in data_outliner_cleaned
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=123)
X_resampled_over, y_resampled_over = ros.fit_resample(data_outliner_cleaned.drop(columns="Diabetes"), data_outliner_cleaned["Diabetes"])

print("Setelah oversampling:")

# Train model with oversampled data
logistic_regression = LogisticRegression(max_iter=10000)
logistic_regression.fit(X_resampled_over, y_resampled_over)
y_pred = logistic_regression.predict(X_val)

accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
# Hyperparameter tuning logistic regression undersampling data train
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'max_iter': [10000],
    'solver': ['lbfgs', 'liblinear']
}

logistic_regression = LogisticRegression()
grid_search = GridSearchCV(estimator=logistic_regression, param_grid=param_grid)
grid_search.fit(X_resampled_under, y_resampled_under)

best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")

y_pred = grid_search.predict(X_val)
accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
# Validation with cross validation score (K = 5)

from sklearn.model_selection import cross_val_score

logistic_regression = LogisticRegression(C=0.1, max_iter=10000, solver='liblinear')
cross_val_scores = cross_val_score(logistic_regression, X_resampled_under, y_resampled_under, cv=5)

print("Cross validation scores:", cross_val_scores)
print("Mean cross validation score:", cross_val_scores.mean())
```
